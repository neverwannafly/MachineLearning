```python
# Learn from training dataset

import numpy
import pandas

from matplotlib import pyplot

file = pandas.read_csv('train_data.csv')
data = numpy.array(file)
n = len(data[0])
m = len(data)

# Augment Cost functions and Gradient Descent methods to include regularization

def get_cost(X, Y, theta, m):
    val = numpy.dot(
        numpy.transpose(theta),numpy.transpose(X)
    ) - numpy.transpose(Y)
    
    cost = 0
    for col in val[0]:
        cost += col*col
        
    # Regularization term
    reg_term = [[0], [0], [1000], [10000], [0], [1000], [10000], [1000], [0]]
    term = numpy.dot(numpy.transpose(numpy.square(theta)), reg_term)
    cost += term[0][0]
        
    return cost / (2*m)

def gradient_decent(X, Y, theta, alpha, m):
    val = numpy.transpose(
        numpy.dot(
            numpy.dot(
                numpy.transpose(theta),numpy.transpose(X)
            ) - numpy.transpose(Y), X
        )
    )
    reg_term = [[0], [0], [100], [1000], [0], [100], [1000], [100], [0]]
    term = numpy.dot(numpy.transpose(theta), reg_term)
    val += term[0][0]
    return theta - ((alpha/m) * val)
    
    
# Get X
X = data[:,0:8]
X = numpy.insert(X, 0, 1, axis=1)

# Get Y
Y = data[:,8]

# Get theta
theta = numpy.zeros((n,1))
```

```python
# Data Visualization
def plot(x_axis, y_axis, title):
    pyplot.scatter(x_axis, y_axis, color='g', marker='x')
    pyplot.title(title)
    pyplot.show()

y_axis = [y for y in Y]
for i in range(1,9):
    x_axis = [x for x in X[:,i]]

    plot(y_axis, x_axis, 'Strength vs Attribute')
    
```

```python
# Data processing: Normalize Data
def preprocess(X):
    mu = numpy.mean(X, axis=0)
    sigma = numpy.std(X, axis=0)
    for i in range(1,n):
        X[:,i] = (X[:,i] - mu[i])/sigma[i]
    return X

X = preprocess(X)
```

```python
# Run Linear Regression
alpha = 0.1
iterations = 1000
history = []

index = 0
while index < iterations:
    history.append(get_cost(X, Y, theta, m))
    theta = gradient_decent(X, Y, theta, alpha, m)
    index += 1

pyplot.plot([i+1 for i in range(iterations)], history)
print(theta)
pyplot.show()
```

```python
# Get Accuracy of model
rms_error = 0

value = numpy.dot(
    numpy.transpose(theta), numpy.transpose(X)
) - numpy.transpose(Y)

for item in value[0]:
    rms_error += item*item
rms_error = (rms_error/m)**0.5

display(rms_error)
```

```python
# Predict Strength and generate output
file_test = pandas.read_csv('test_data.csv')
data_test = numpy.array(file_test)
n_test = len(data_test[0])
m_test = len(data_test)

# Get test X
X_test = data_test[:,0:8]; X_test = numpy.insert(X_test, 0, 1, axis=1)

# Process test data
X_test = preprocess(X_test)

prediction = numpy.transpose(numpy.dot(
    numpy.transpose(theta), numpy.transpose(X_test)
))

pandas.DataFrame(prediction).to_csv('output.csv')
```
